```python
#febinacci series
def fibinacci(n):
  if(n==1):
   return 0
  elif(n==2):
    return 1
  else:
    return fibinacci(n-1)+fibinacci(n-2)
fibinacci(5)
```

```python
# factorial series
def fact(n):
  if(n==0):
    return 1
  elif(n==1):
    return 1
  else:
    result= n*fact(n-1)
  return result
print(fact(4))
```
